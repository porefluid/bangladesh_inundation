```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_dir = '/Users/corey/OneDrive - City University of New York/CHART/inundation/results/unions'
fig_dir = '/Users/corey/Documents/GitHub/bangladesh_inundation/figures'
```

```python
paths = []

for i in os.listdir(data_dir):
    paths.append(os.path.join(data_dir, i))
```

```python
df = pd.DataFrame(columns=['datetimes', 'VV'])
for i in paths:
    df = pd.concat([df, pd.read_csv(i, parse_dates=['datetimes'])])
```

```python
df = df.set_index('datetimes').sort_index()
```

```python
df.plot(marker='.', linewidth=0)
```

```python
df_30 = df.resample('30D').sum()

df_30 = (df_30 * 100 * 100) / 1e3

fig = df_30.plot(title = 'Monthly area of "inundated" pixels')

fig.set_ylabel('count')
fig.legend(['pixel count'])
plt.savefig(os.path.join(fig_dir, 'example_results.png'), dpi=300)
```
